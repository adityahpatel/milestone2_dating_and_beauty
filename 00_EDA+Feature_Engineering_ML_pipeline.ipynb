PointCloud is basically a 3D coordinate system, in which points are usually defined by X, Y, and Z coordinates. Used to describe external surface of an object.

# Reviewing Facial Landmarks

```python
# this is a helper function to visually review data
def img_map(filename, annotate=False):
    """
    PARAMS:
    filename: ie AM1
    annotate: Boolean to numerically label pts
    
    OUTPUT:
    plot of image with landmarks
    """
    # repeat for asian male
    pts_path = 'data/SCUT-FBP5500_v2/landmark_txt/' + filename + '.txt'
    img_path = 'data/SCUT-FBP5500_v2/Images/' + filename + '.jpg'
    
    # read landmarks 
    with open(pts_path,'r') as f:
        lines = f.readlines()
        points = [l.split() for l in lines]
        p_float = [(float(c[0]),float(c[1])) for c in points]

    # create xs and ys
    xs = [c[0] for c in p_float]
    ys = [c[1] for c in p_float]
    
    # plot figure
    plt.figure(figsize=(10,10))
    plt.imshow(plt.imread(img_path))
    plt.scatter(xs, ys, marker="o", s=3, c='r')
    
    # label landmarks
    if annotate:
        for i, c in enumerate(p_float):
            plt.text(int(c[0]), int(c[1]), str(i), color="g", fontsize=12)

    # resize figure for exporting
    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(10.5, 10.5, forward=True)
    plt.title(filename)
    plt.show()
    return
```

A good keypoint is one which is invariant to all these conditions. It is a Scale-invariant feature transform (SIFT) processed form of a landmark/keypoint.

## A Keypoint has the following properties:
* x, y: Keypoint coordinates in the image it was extracted from. 
* response: “Strength” of a feature. A higher score means a higher local contrast.
* angle: Orientation angle (in radians)
* scale: Radius of the support region (in pixels)
* level: Scale level from which the feature was extracted. Starts at zero.

```python
# skipped the above two methods for a filtering technique instead
def generate_features_df(filename, directory, df_orig, shape=True):
    """
    This function generates all the Supervised ML features we will use pixel by pixel.
    
    Dependency: 
    from skimage.filters import roberts, sobel, scharr, prewitt
    from scipy import ndimage as nd
    
    parmeters: 
    filename = string e.g CM1
    directory = root folder of data e.g 'data/SCUT-FBP5500_v2'
    df_orig = pd.Dataframe()
    if shape then reshape to pixel form
    """
    df = pd.DataFrame()
    df['Filename'] = [filename + '.jpg']
    # grayscale is used so that artifacts don't appear when filtering
    img = cv2.cvtColor(cv2.imread(directory+ '/Images/'+ filename +'.jpg'), cv2.COLOR_BGR2GRAY) 
    
    # Gender & Race encoding
    txt = os.path.splitext(filename)[0]
    if txt[1] == 'M' :
        df['male'] = 1
    if txt[0] == 'A':
        df['asian'] = 1
    
    # Save original image pixels into a data frame. This is our Feature #1.
    df['Orig_img'] = [img]
    img2 = img.reshape(-1)
    
    landmarks_path = directory +'/landmark_txt/' + filename + '.txt'
    try:
        with open(landmarks_path ,'r') as f:
            lines = f.readlines()
            points = [l.split() for l in lines]
            p_float = [(float(c[0]),float(c[1])) for c in points]
        df['landmarks'] = [np.array(p_float)]
    except:
        print(f'txt file does not exist for {filename}')
        df['landmarks'] = [np.nan]
    

    # ORB object for Fast open source SIFT
    ORB_object = cv2.ORB_create(nfeatures=300)
    
    # detect and compute SIFT keypoints
    keypoints, descriptors = ORB_object.detectAndCompute(img, None)
    df['orb_kp'] = [keypoints]
    df['orb_dec'] = [descriptors]
    
    ### EDGE DETECTION ###
    
    # Generate Gabor features manually because sci-kit gabor() breaks machine
    num = 1  # To count numbers up in order to give Gabor features a label in the data frame
    kernels = []
    for theta in range(2):   # Define number of thetas
        theta = theta / 4. * np.pi
        for sigma in (1, 3):  # Sigma with 1 and 3
            for lamda in np.arange(0, np.pi, np.pi / 4):   # Range of wavelengths
                for gamma in (0.05, 0.5):   # Gamma values of 0.05 and 0.5
                
                    
                    gabor_label = 'Gabor' + str(num)  # Label Gabor columns as Gabor1, Gabor2, etc.
                    ksize=9
                    kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, 0, ktype=cv2.CV_32F)    
                    kernels.append(kernel)
                    
                    # Now filter the image and add values to a new column 
                    fimg = cv2.filter2D(img2, cv2.CV_8UC3, kernel)
                    if shape:
                        fimg = fimg.reshape(-1)
                        df[gabor_label] = fimg
                    df[gabor_label] = [fimg]  # Labels columns as Gabor1, Gabor2, etc.
                    # print(gabor_label, ': theta=', theta, ': sigma=', sigma, ': lamda=', lamda, ': gamma=', gamma)
                    num += 1  # Increment for gabor column label              
      
    # To visualize any of these filters use e.g imshow(edges_roberts, cmap='gray')            
    # CANNY EDGE
    edges = cv2.Canny(img, 100,200)   #Image, min and max values
    if shape:
        edges = edges.reshape(-1)
        df['Canny Edge'] = edges
    df['Canny Edge'] = [edges] #Add column to original dataframe

    # ROBERTS EDGE
    edge_roberts = roberts(img)
    if shape:
        edge_roberts = edge_roberts.reshape(-1)
        df['Roberts'] = edge_roberts
    df['Roberts'] = [edge_roberts]

    # SOBEL EDGE
    edge_sobel = sobel(img)
    if shape:
        edge_sobel = edge_sobel.reshape(-1)
        df['Sobel'] = edge_sobel
    df['Sobel'] = [edge_sobel]

    # SCHARR EDGE
    edge_scharr = scharr(img)
    if shape:
        edge_scharr = edge_scharr.reshape(-1)
        df['Scharr'] = edge_scharr
    df['Scharr'] = [edge_scharr]

    # PREWITT EDGE
    edge_prewitt = prewitt(img)
    if shape:
        edge_prewitt1= edge_prewitt.reshape(-1)
        df['Prewitt'] = edge_prewitt
    df['Prewitt'] = [edge_prewitt]

    # GAUSSIAN with sigma=3
    gaussian_img = nd.gaussian_filter(img, sigma=3)
    if shape:
        gaussian_img = gaussian_img.reshape(-1)
        df['Gaussian s3'] = gaussian_img
    df['Gaussian s3'] = [gaussian_img]

    # GAUSSIAN with sigma=7
    gaussian_img2 = nd.gaussian_filter(img, sigma=7)
    if shape:
        gaussian_img2 = gaussian_img2.reshape(-1)
        df['Gaussian s7'] = gaussian_img2
    df['Gaussian s7'] = [gaussian_img2]

    # MEDIAN with sigma=3
    median_img = nd.median_filter(img, size=3)
    if shape:
        median_img = median_img.reshape(-1)
        df['Median s3'] = median_img
    df['Median s3'] = [median_img]

    # VARIANCE with size=3
    variance_img = nd.generic_filter(img, np.var, size=3)
    if shape:
        variance_img = variance_img.reshape(-1)
        df['Variance s3'] = variance_img
    df['Variance s3'] = [variance_img]  #Add column to original dataframe
    
    return pd.concat([df_orig, df])
    
```
